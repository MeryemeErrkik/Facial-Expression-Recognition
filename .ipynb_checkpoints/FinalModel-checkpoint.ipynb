# Preprocessing#

Since we are not interested with the color of the images and with only the face regardless of the position of the body or the background we will convert the images to gray and crop the faces using HAAR cascade

**Converting to grayscale**

```python
import cv2
from os import listdir,makedirs
from os.path import isfile,join

def gray(path):
    dstpath=path
    try:
        makedirs(dstpath)
    except:
        print ("Directory already exist, images will be written in asme folder")

    # Folder won't used
    files = [f for f in listdir(path) if isfile(join(path,f))] 

    for image in files:
        try:
            img = cv2.imread(os.path.join(path,image))
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            dstPath = join(dstpath,image)
            cv2.imwrite(dstPath,gray)
        except:
            print ("{} is not converted".format(image))
```

```python
from glob import glob
subs=glob("./data/*")
for emo in subs:
    gray(emo)
```

**Detecting and cropping faces:**

```python
import cv2
from os import listdir,makedirs
from os.path import isfile,join

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 
def crop_face(path):
    dstpath=path
    try:
        makedirs(dstpath)
    except:
        print ("Directory already exist, images will be written in asme folder")

    # Folder won't used
    files = [f for f in listdir(path) if isfile(join(path,f))] 

    for image in files:
        try:
            img = cv2.imread(os.path.join(path,image))
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)
            for (x,y,w,h) in faces:
                face_clip = img[y:y+h, x:x+w]  #cropping the face in image
                dstPath = join(dstpath,image)
                cv2.imwrite(dstPath, cv2.resize(face_clip, (224, 224)))  #resizing image then saving it

            
        except:
            print ("{} is not converted".format(image))

from glob import glob
subs=glob("./data/*")
for emo in subs:
    crop_face(emo)




```

# Our Model#

We used transfer learning for our task and we choosed the MobileNet since it needs less computation power to run or apply transfer learning to than other architecture.


```python
base_model=MobileNet(weights='imagenet',include_top=False,input_shape=(224, 224, 3)) #imports the mobilenet model and discards the last 1000 neuron layer.
base_model.summary()
```

```python
x=base_model.output
x=GlobalAveragePooling2D()(x)
x=Dense(1024,activation='relu')(x) 
x=Dense(1024,activation='relu')(x) 
x=Dense(512,activation='relu')(x) 
preds=Dense(7,activation='softmax')(x) 
model=Model(inputs=base_model.input,outputs=preds)
```

We will use pre-trained weights as the model has been trained already on the Imagenet dataset. We ensure all the weights are non-trainable. We will only train the last few dense layers.

```python
for layer in model.layers[:20]:
    layer.trainable=False
for layer in model.layers[20:]:
    layer.trainable=True
```

We will use **Tensorboard** to visualize The Loss function

```python
log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
# Specify the callback object
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

Now lets load the training data into the ImageDataGenerator. Specify path, and it automatically sends the data for training in batches, simplifying the code.

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])


train_datagen = ImageDataGenerator(rescale=1./255,
    validation_split=0.2) # set validation split

train_generator = train_datagen.flow_from_directory(
    './data/',
    target_size=(224,224),
    color_mode='rgb',
    batch_size=64,
    class_mode='categorical',
    subset='training',
    shuffle=True) # set as training data

validation_generator = train_datagen.flow_from_directory(
        './data/',
        target_size=(224,224),
        color_mode='rgb',
        batch_size=64,
        class_mode='categorical',
        subset='validation',
        shuffle=True) # set as validation data
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch = train_generator.n//train_generator.batch_size,
    validation_data = validation_generator, 
    validation_steps = validation_generator.n // validation_generator.batch_size,
    epochs = 3, callbacks=[tensorboard_callback])
```
